````python
def get_input():
  print("Enter num, p, d: ")
  num = int(input())
  p = int(input())
  d = int(input())
  return num, p, d

def add_zeros(num, p):
  #print("Debug: adding zeros to: ", num)
  str_num = str(num)
  t = "0"* (p -1)
  t_str_num = str_num[:len(str_num) - p + 1] + t
  num = int(t_str_num)
  print("Debug: Final num: ", num)

def transform_one(num, p, d):
  #Transform
  print("Debug: first transformation")
  n_p = int(str(num)[-p])
  n_t = abs(n_p + d)
  n_d = n_t%10
  str_num = str(num)
  t_str_num = str_num[:len(str_num) - p] + str(n_d) + str_num[len(str_num) - p + 1: ]
  num = int(t_str_num)
  add_zeros(num, p)

def transform_two(num, p, d):
  #Transform
  #print("Debug: second transformation")
  n_p = int(str(num)[-p])
  n_t = abs(n_p - d)
  n_d = int(str(n_t)[0])
  str_num = str(num)
  t_str_num = str_num[:len(str_num) - p] + str(n_d) + str_num[len(str_num) - p + 1: ]
  num = int(t_str_num)
  add_zeros(num, p)

def check_p(num, p, d):
  #print("Debug: check p")
  num_str = str(num)
  if int(num_str[-p]) >= 0 and int(num_str[-p]) <= 4:
    transform_one(num, p, d)
  elif int(num_str[-p]) >= 5 and int(num_str[-p]) <= 9:
    transform_two(num, p, d)
  else:
    return False

def main():
  num, p, d = get_input()
  check_p(num, p, d)

main()
````

````python
class Cow:
    def __init__(self, cow_id):
        self.cow_id = cow_id
        self.yields = []

    def record_yield(self, milk_yield):
        self.yields.append(milk_yield)

    def total_yield(self):
        return sum(self.yields)

    def low_yield_days(self):
        return len([yield_value for yield_value in self.yields if yield_value < 12])


class Herd:
    def __init__(self, num_cows):
        self.cows = [Cow(f"Cow-{str(i).zfill(3)}") for i in range(1, num_cows + 1)]

    def record_milk_yield(self):
        for cow in self.cows:
            print(f"Enter milk yields for {cow.cow_id} (14 milking sessions - twice a day for 7 days):")
            for i in range(1, 3):
                while True:
                    try:
                        yield_value = float(input(f"Milking session {i}: "))
                        if yield_value < 0:
                            raise ValueError("Yield can't be negative.")
                        cow.record_yield(round(yield_value, 1))
                        break
                    except ValueError as e:
                        print(e)

    def calculate_statistics(self):
        total_milk = 0
        for cow in self.cows:
            weekly_total = cow.total_yield()
            total_milk += weekly_total
            print(f"{cow.cow_id}: Total weekly yield = {weekly_total:.1f} litres")

        # Total and average yield
        avg_yield = total_milk / len(self.cows)
        print(f"\nTotal milk yield for the herd: {total_milk:.1f} litres")
        print(f"Average milk yield per cow: {avg_yield:.1f} litres")

    def identify_best_and_low_producers(self):
        best_cow = max(self.cows, key=lambda cow: cow.total_yield())
        print(f"\nCow with the highest yield: {best_cow.cow_id} with {best_cow.total_yield():.1f} litres")

        print("\nCows with less than 12 litres on four or more days:")
        low_producers = [cow for cow in self.cows if cow.low_yield_days() >= 4]
        if low_producers:
            for cow in low_producers:
                print(f"{cow.cow_id}: Total weekly yield = {cow.total_yield():.1f} litres")
        else:
            print("None")


def run():
    herd_size = int(input("Enter the number of cows in the herd: "))
    herd = Herd(herd_size)

    herd.record_milk_yield()

    print("\nCalculating weekly statistics...")
    herd.calculate_statistics()

    herd.identify_best_and_low_producers()

run()
````
